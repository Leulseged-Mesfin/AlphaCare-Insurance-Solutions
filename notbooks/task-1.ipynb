```python
# 3. Multivariate Analysis: Scatter plot for each PostalCode using FacetGrid

# Limit the number of unique PostalCodes
top_postal_codes = df['PostalCode'].value_counts().index[:9]  # Top 9 most frequent PostalCodes
df_subset = df[df['PostalCode'].isin(top_postal_codes)]

# Create a FacetGrid with 3 columns and wrap the rest into rows
g = sns.FacetGrid(df_subset, col="PostalCode", col_wrap=3, height=5)  # Make each plot bigger with height=5
g.map(sns.scatterplot, "TotalPremium", "TotalClaims")
g.add_legend()

# Show the plot
plt.show()
```

```python
# top_postal_codes = df['PostalCode'].value_counts().index[:3]  # Top 3 most frequent PostalCodes
# top_vehicle_makes = df['make'].value_counts().index[:3]       # Top 3 most frequent vehicle makes

# df_subset = df[df['PostalCode'].isin(top_postal_codes) & df['make'].isin(top_vehicle_makes)]
# # df_subset = df[df['make'].isin(top_vehicle_makes)]

# # Create a FacetGrid to compare TotalPremium across PostalCodes for different vehicle makes
# g = sns.FacetGrid(df_subset, col="PostalCode", height=8, aspect=2)
# g.map(sns.boxplot, "make", "TotalPremium")
# g.add_legend()

# plt.show()
# print(top_vehicle_makes)
# print(df_subset['make'].value_counts())




top_postal_codes = df['PostalCode'].value_counts().index[:3]  # Top 3 most frequent PostalCodes
top_vehicle_makes = df['make'].value_counts().index[:3]       # Top 3 most frequent vehicle makes

df_subset = df[df['PostalCode'].isin(top_postal_codes) & df['make'].isin(top_vehicle_makes)]

# Plot each PostalCode separately
for postal_code in top_postal_codes:
    plt.figure(figsize=(14, 10))  # Adjust size as needed
    subset = df_subset[df_subset['PostalCode'] == postal_code]
    
    # Compute average TotalPremium for each make
    avg_premium = subset.groupby('make')['TotalPremium'].mean().reset_index()
    
    # Plot bar chart
    sns.barplot(x='make', y='TotalPremium', data=avg_premium)
    plt.title(f'Average TotalPremium by Vehicle Make for PostalCode = {postal_code}')
    plt.xlabel('Vehicle Make')
    plt.ylabel('Average Total Premium')
    plt.xticks(rotation=45)  # Rotate x labels if needed
    plt.tight_layout()
    plt.show()
```
